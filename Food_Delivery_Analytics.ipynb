## Stage 2 – Data Cleaning and Pre-processing

Handling Missing Values in Categorical and Numerical Columns

```python
# Numerical column
df['OrderValue'].fillna(df['OrderValue'].mean(), inplace=True)
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['CustomerRating'].fillna(df['CustomerRating'].mean(), inplace=True)


```

```python
df
```

```python
# Categorical column
df['City']=df['City'].fillna("Unknown")
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Converting datatypes
df['Quantity'] = df['Quantity'].astype(int)
df['CustomerRating'] = df['CustomerRating'].astype(int)
```

## Stage 3 – EDA and Visualizations



# 1. Orders by City (Count Plot)

```python
plt.figure(figsize=(8,5))
sns.countplot(x="City", data=df, palette="Set2")
plt.title("Number of Orders by City")
plt.xlabel("City")
plt.ylabel("Order Count")
plt.show()
```

#Interpretation:

This visualization compares order volumes across cities, showing Delhi as the top contributor, Mumbai close behind, and other cities with strong but slightly lower demand.

# Revenue by Restaurant (Bar Plot)

```python
plt.figure(figsize=(10,5))
df.groupby("Restaurant")["TotalAmount"].sum().sort_values().plot(kind="bar", color="skyblue")
plt.title("Total Revenue by Restaurant")
plt.ylabel("Revenue")
plt.show()
```

# Interpretation of Bar Plot:

 The chart compares restaurant revenues side by side, showing that Healthy Bowl generates the most revenue, while the rest are close behind with only small differences.

# Payment Method Split (Pie Chart)

```python
df["PaymentMethod"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

# Interpretation of Pie Chart:

Displays customer preferences for payment. If UPI dominates, it means digital payments are most popular. A large share of cash indicates reliance on traditional methods.

# Top Food Items Ordered (Bar Plot)


```python
plt.figure(figsize=(10,5))
df["FoodItem"].value_counts().sort_values(ascending=False).plot(kind="bar", color="purple")
plt.title("Top Ordered Food Items")
plt.ylabel("Order Count")
plt.show()

```

# Interpretation of Box Plot:

This chart shows customer preferences: Wrap and Biryani dominate demand, Coffee and Sandwich are strong contenders, and Taco and Pasta have lower popularity. It helps identify bestsellers vs. underperformers for menu planning and promotions.

# Delivery Time Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
plt.hist(df["DeliveryTime_Minutes"], bins=30, color="orange", edgecolor="black")
plt.title("Delivery Time Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()

```

# Interpretation of Histogram:

Shows how long deliveries usually take. If most orders fall under 40 minutes, service is efficient. A long tail of 60+ minutes highlights delays that may hurt customer satisfaction.

# Revenue Over Time (Line Chart)

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"], errors="coerce")
daily_revenue = df.groupby("OrderDate")["TotalAmount"].sum()
plt.figure(figsize=(12,5))
plt.plot(daily_revenue.index, daily_revenue.values, color="green")
plt.title("Revenue Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

# Interpretation of Line Chart:

Reveals seasonal or monthly peaks. For example, spikes around weekends or festivals indicate demand surges. Flat or declining lines suggest periods of low activity.

# Customer Ratings by City (Box Plot)

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="City", y="CustomerRating", data=df, palette="Set3")
plt.title("Customer Ratings by City")
plt.show()
```

# Interpretation of Bar Plot:

Compares satisfaction across cities. Cities with higher median ratings (like Bangalore or Hyderabad) show happier customers. Wider boxes or many outliers suggest inconsistent experiences.

# Order Value vs Delivery Time (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="DeliveryTime_Minutes", y="OrderValue", hue="City", data=df)
plt.title("Order Value vs Delivery Time")
plt.show()
```

# Interpretation of Scatter Plot:

Tests if higher-value orders take longer. If points cluster upward with longer times, premium orders may be slowing delivery. A flat spread means order value doesn’t affect speed.

# Multivariate Analysis (Three+ Variables)

# Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Quantity","OrderValue","TotalAmount","DeliveryTime_Minutes","CustomerRating"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Interpretation of Correlation Heatmap:

Highlights relationships between variables. Strong correlation between OrderValue and TotalAmount is expected. Weak or negative correlations (like Delivery Time vs Rating) show where improvements matter.
